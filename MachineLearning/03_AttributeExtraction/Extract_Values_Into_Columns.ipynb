```python
import json
import pandas as pd
import re
```

```python
#Load JSON Data
json_data = None
json_data = 'AmazonElectronics_Labeled.json'
data = pd.read_json(json_data, orient='records')
```

### Get Product Dimension from nested JSON

```python
#Write nested product dimensions in new column "size_new"
data['size']="-"

for i in range(data.shape[0]):
    
    if pd.isnull(data.iloc[i].technical_details__k_v_pairs):
        print('No dimension row - {}'.format(i))
    elif type(data.iloc[i].technical_details__k_v_pairs) == str:
        print('String Detected  - No dimension row - {}'.format(i))
    else:
        dict_keys = list(data.iloc[i].technical_details__k_v_pairs.keys())               
        if 'Product_Dimensions' in dict_keys:
            dim=data.iloc[i].technical_details__k_v_pairs['Product_Dimensions']
            #store nested value in new column:
            data['size'][i]=dim
        else:
            if pd.isnull(data.iloc[i].product_details__k_v_pairs):
                print('No dimension in product details row - {}'.format(i))
            elif type(data.iloc[i].product_details__k_v_pairs) == str or type(data.iloc[i].product_details__k_v_pairs) == float:
                print('String Detected at product details - No dimension row - {}'.format(i))
            else:
                dict_keys = list(data.iloc[i].product_details__k_v_pairs.keys())                
                if 'Product_Dimensions' == dict_keys:
                    dim=data.iloc[i].product_details__k_v_pairs['Product_Dimensions']
                    #store nested value in new column:
                    data['size'][i]=dim
                    
```

## regex

```python
#extract length from product dimensions into new column
data['length_cm'] = data['size'].str.extract('(.*)(?= x .* x .*)', expand=True)
```

```python
#extract width from product dimensions into new column
data['width_cm'] = data['size'].str.extract('(?<=x )(.*)(?= x .*)', expand=True)
```

```python
#extract height from product dimensions into new column
data['height_cm'] = data['size'].str.extract('x.*?x(.*)(?=\s+cm)', expand=True)
```

```python
#extract weight in grams from weight and save in new column
data['weight_g'] = data['weight'].str.extract('(.*)(?= Grams| g| G)', expand=True)
```

```python
#Extract weight in kilograms from weight and save in new column
data['weight_kg'] = data['weight'].str.extract('(.*)(?= kilograms| Kg| kg)', expand=True)
```

```python
#save file with extracted data
json_object = json.dumps([row.dropna().to_dict() for index,row in data.iterrows()],indent=4)

with open('AmazonElectronics_labeled_extracted.json', 'w') as outfile:
    outfile.write(json_object) 
```
